<font style="font-size:2.5em; color:red"> 2. Mediciones de distancia</font> 

<p>Considere dos puntos, $A$ y $B$ (Fig. 1), donde $X_1$ y $Y_1$ son las coordenadas del punto $A$, y donde $X_2$ y $Y_2$ son las coordenadas del punto $B$. La distancia entre el punto $A$ y el punto $B$ se podria estimar a través de movimientos verticales y horizonales (Fig.1, líneas moradas) para llegar de un punto a otro, o mediante un trazo directo entre ambos puntos (Fig. 1, línea verde). Algunos métodos comunmente empleados para medir distancias son: la distancia euclidea, método Manhattan y la distancia Minkowsky.</p>

```python
pA = [1,1]
pB = [2,2]

plt.scatter(pA[0], pA[1])
plt.scatter(pB[0], pB[1])
plt.plot([1,2],[1,2], color="green")
plt.annotate("Punto A", (1.05,1.05))
plt.annotate("Punto B", (1.8,2))
plt.hlines(1,1,2, color="purple")
plt.vlines(2,1,2, color="purple")
plt.figtext(0.5, 0.01, "Figura 1. Distancias recorrifas entre el punto A y el punto B", wrap=True, horizontalalignment='center', fontsize=12)
plt.show()
```

<font style="font-size:2em; color:green"> 2.1. Distancia euclidea</font> 

<p>La distancia ($S$) euclidea es estimada mediante el teorema de Pitágoras, el cual se define como la raíz cuadrada de la suma de las diferencias cuadradas entre las coordenadas de dos puntos (Ec.1):</p>

$S = \sqrt{ \left(X_2 - X_1 \right)^{2} + \left(Y_2 - Y_1 \right)^{2} } \tag{1}$

<font style="font-size:2em; color:green"> 2.2. Distancia por método Manhattan</font> 

<p>La distancia ($S$) Manhattan se define como la suma del valor absoluto de la diferencia vertical ($\mid Y_2 - Y_1 \mid$) y horizontal ($\mid X_2 - X_1 \mid$) de las coordenadas de dos puntos (Ec.2):</p>

$S = \mid Y_2 - Y_1 \mid  + \mid X_2 - X_1 \mid \tag{2}$

<font style="font-size:2em; color:green"> 2.3. Distancia Minkowsky</font> 

<p>El método de medición de la distancia ($S$) Minkowsky es una generalización de la distancia euclidea y de Manhattan. En esta generalización se emplea un parámetro ($p$), el cual puede tomar valores de 1 (para Manhattan) o 2 (para euclidea) (Ec. 3): </p>

$S = \left[ \left(X_2 - X_1 \right)^{p} + \left(Y_2 - Y_1 \right)^{p} \right]^{1/p} \tag{3}$
