```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
prec = np.random.rand(12,4,5)*100
```

```python
prec
```

```python
temp = np.random.rand(12,4,5)*25
```

```python
temp
```

```python
fechas = pd.date_range("2018-01-01", periods=12, freq="M")
lons = np.linspace(-76, -70, 5)
lats = np.linspace(25, 55, 4)
```

```python
prec_xr = xr.DataArray(prec, coords=[fechas,lats,lons],dims=('time','lat','lon'))
```

```python
prec_xr
```

```python
temp_xr = xr.DataArray(temp, coords=[fechas,lats,lons],dims=('time','lat','lon'))
```

```python
temp_xr
```

```python
temp_xr.attrs['unidades'] = 'C'
prec_xr.attrs['unidades'] = 'mm'
```

```python
prec_xr
```

```python
datos = xr.Dataset(data_vars={'prec':prec_xr,'temp':temp_xr})
```

```python
datos
```

```python
datos.temp
```

```python
datos["prec"]
```

```python
datos.sel(time="2018-01-31")
```

```python
datos.sel(time="2018-01-31",lat=25)
```

```python
datos.sel(time="2018-02-01",lat=25,method="nearest")
```

```python
from datetime import timedelta
datos.sel(time="2018-08-30", method="nearest", tolerance=timedelta(days=2))
```

```python
datos.sel(time=slice("2018-01-01", "2018-07-03"))
```

```python
datos.isel(lat=0)
```
